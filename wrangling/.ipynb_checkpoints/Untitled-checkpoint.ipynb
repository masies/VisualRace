```python
import pandas as pd

def makeDataframe(path,name,raceName):
    df = pd.read_csv(path, encoding= 'unicode_escape') 
    df = df.dropna(axis=1, how='all')

    # data = data[{"Time","Distance","LateralAcc","InlineAcc","VerticalAcc","RPM","Throttle","GPS_Speed","GPS_LatAcc","GPS_LonAcc","GPS_Latitude","GPS_Longitude","GPS_Elevation"}]
    df['coordinates'] = df.apply(lambda row: [row['GPS_Longitude'], row['GPS_Latitude']], axis=1)

    df.insert(loc=0, column='Driver', value=name)
    df.insert(loc=0, column='Race', value=raceName)
    return df

def InsertLapsAndTimeStamp(df):
    lapIndexes = []
    lapIndex = 1
    for i in range(len(df)-1):
        if (df.loc[i, "Time"] < df.loc[i+1, "Time"]):      
            lapIndexes.append(lapIndex)
        else:
            lapIndex += 1
            lapIndexes.append(lapIndex-1)
    lapIndexes.append(lapIndex)
 
    df["Lap"] = lapIndexes

    times = [x/10 for x in range(0,len(df))]
    df["TimeStamp"] = times
    return df
    
    
dataMachado = makeDataframe("./data/Machado.csv","Machado","Varano")
dataSolfaroli = makeDataframe("./data/Solfaroli.csv","Solfaroli","Varano")
dataBini = makeDataframe("./data/Bini.csv","Bini","Varano")
dataBighetti = makeDataframe("./data/Bighetti.csv","Bighetti","Varano")
dataAntongini = makeDataframe("./data/Antongini.csv","Antongini","Varano")
dataRizzi = makeDataframe("./data/Rizzi.csv","Rizzi","Varano")

dataMachado =  InsertLapsAndTimeStamp(dataMachado)
dataSolfaroli = InsertLapsAndTimeStamp(dataSolfaroli)
dataBini = InsertLapsAndTimeStamp(dataBini)
dataBighetti =  InsertLapsAndTimeStamp(dataBighetti)
dataAntongini = InsertLapsAndTimeStamp(dataAntongini)
dataRizzi = InsertLapsAndTimeStamp(dataRizzi)

dataSolfaroliNuvola = makeDataframe("./data/Solfaroli_Nuvola.csv","Solfaroli","Nuvolari")
dataMachadoNuvola = makeDataframe("./data/Machado_Nuvola.csv","Machado","Nuvolari")
dataBiniNuvola = makeDataframe("./data/Bini_Nuvola.csv","Bini","Nuvolari")
dataRizziNuvola = makeDataframe("./data/Rizzi_Nuvola.csv","Rizzi","Nuvolari")

dataSolfaroliNuvola = InsertLapsAndTimeStamp(dataSolfaroliNuvola)
dataMachadoNuvola =  InsertLapsAndTimeStamp(dataMachadoNuvola)
dataBiniNuvola = InsertLapsAndTimeStamp(dataBiniNuvola)
dataRizziNuvola = InsertLapsAndTimeStamp(dataRizziNuvola)

all_pilots = pd.concat([dataMachado, dataSolfaroli,dataBini,dataRizzi,dataAntongini,
                        dataBighetti,dataSolfaroliNuvola,dataMachadoNuvola,dataBiniNuvola,dataRizziNuvola], ignore_index=True)
all_pilots.to_json('./res/data_all.json', orient="records", lines=True)



```
